###### Image processing using NumPy

```python
import numpy as np
from matplotlib import pyplot as plt
test_image= plt.imread('./files/ancient.jpg')
type(test_image)
```

ndarray is a multidimensional array

```python
vertically_flipped= np.flipud(test_image)
def show_image(image, title="image",cmap_type="gray"):
    plt.imshow(image,cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
show_image(vertically_flipped,'vertically flipped image')
```

```python
horizontally_flipped= np.fliplr(test_image)
show_image(horizontally_flipped,'horizontally flipped')
```

```python
#histogram
blue= test_image[:,:,2]
plt.hist(blue.ravel(),bins=256)
plt.title("blue hisogram of image")
plt.show()
```

```python
#red histogram
red=test_image[:,:,0]
plt.hist(red.ravel(),bins=256)
plt.title("red histogram")
plt.show()
```

```python
#green histogram
green=test_image[:,:,1]
plt.hist(green.ravel(),bins=256)
plt.title("green histogram")
plt.show()
```

```python

from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu
from skimage import color 
image=plt.imread('./files/threshold.jpg')

image_gray = color.rgb2gray(image)

thresh = threshold_otsu(image_gray)


binary = image_gray >  thresh
def show_image(image, title="image",cmap_type="gray"):
    plt.imshow(image,cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

show_image(image,'original image')
show_image(binary, 'Binary image')
```

```python
#global & Local thresholding
image=plt.imread('./files/blurry_background.png')
show_image(image,"original")
from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu
from skimage import color 
image_gray = color.rgb2gray(image)
#show_image(image_gray,"grayscale")

thresh = threshold_otsu(image_gray)

binary = image_gray >  thresh
show_image(binary, 'Binary image')


```

```python
fruits_image=plt.imread('./files/fruits.jpg')
show_image(fruits_image,"original before thresholding")
# Import the try all function
from skimage.filters import try_all_threshold

# Import the rgb to gray convertor function 
from skimage.filters import threshold_otsu 
from skimage.color import rgb2gray 

# Turn the fruits_image to grayscale
grayscale = rgb2gray(fruits_image)

# Use the try all method on the resulting grayscale image
fig, ax = try_all_threshold(grayscale, verbose=False)

# Show the resulting plots
plt.show()
```

```python
#binarized Image
apply=plt.imread('./files/apply.jpg')
show_image(apply,'original image')

from skimage.filters  import threshold_otsu
from skimage.color import rgb2gray

gray_image = rgb2gray(apply)

thresh =threshold_otsu(gray_image)

binary_image = gray_image > thresh

show_image(binary_image, 'Binarized image')
```
